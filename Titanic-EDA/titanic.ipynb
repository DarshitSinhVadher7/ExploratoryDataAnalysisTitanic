```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)

df.to_csv("titanic.csv", index=False)

print("Downloaded", df.shape)
```

```python
display(df.head())
display(df.shape)
display(df.dtypes)
display(df.describe())
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df.drop(columns=['Cabin'], inplace=True, errors='ignore')
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
# //UNIVARIARIATE ANALYSIS
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(3,3))

sns.countplot(x='Survived', data=df)

plt.title('Survival count')

plt.xticks([0,1], ['Died','Survived'])

plt.show()
```

```python
df['Pclass'].value_counts().sort_index().plot(kind='bar')

plt.title('Passengers by class')

plt.xlabel('Class', fontweight='bold')

plt.xticks([0,1,2], ['First','Second','Third'], rotation=0)

plt.show()
```

```python
df['Age'].hist(
    bins=10,
    color='steelblue',
    edgecolor='white'
)

plt.title('Age distribution')

plt.xlabel('Age')

plt.ylabel('Count')

plt.show()
```

```python
# BIRVARIATE ANALYSIS
```

```python
plt.figure(figsize=(3,3))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival rate by sex')
plt.ylabel('Survival rate')
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.barplot(x='Pclass', y='Survived', data=df, errorbar=None)
plt.title('Survival rate by class')
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.xticks([0,1], ['Died','Survived'])
plt.title('Age distribution by survival')
plt.show()
```

```python
sns.histplot(data=df, x='Fare',
  hue='Survived', bins=20, kde=True)
plt.title('Fare distribution by survival')
plt.xlim(0, 200)
plt.show()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=[0, 12, 18, 35, 60, 80],
    labels=['Child', 'Teen', 'Adult', 'Middle-aged', 'Senior']
)

df['Title'] = df['Name'].str.extract(
    r' ([A-Za-z]+)\.',
    expand=False
)
df['Title'].value_counts()
```
